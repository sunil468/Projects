## Number Of Matches Won By Each Team In The World Cup:

```python
data.winner.value_counts().plot.bar(width = 0.75)
plt.title("Number Of Matches Won By Each Team")
plt.xlabel("Winner")
plt.ylabel("count")
plt.xlim(1,30)
plt.ylim(1,5)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.hist(data['winner'],color = 'red', bins = 50) #data['winner'].astype(int)
```

### As Englond won the t20 worldcup 2022, Englond won 5 matches and both India and Pakistan won 4 matches

## Number Of Matches Won By Batting First Or Second

```python
won_by = data["won by"].value_counts()
for category, count in won_by.items():
    print(f'{category}: {count}')
plt.pie(won_by, labels = won_by.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Number of Matches Won By Runs Or Wickets')
plt.show()
```

### So in the t20 world cup 2022, 16 matches were won by batting first, and 13 matches were won by chasing.

## The Toss Decisions By Teams In The World Cup:

```python
toss = data["toss decision"].value_counts()
for category, count in toss.items():
    print(f'{category}: {count}')
plt.pie(toss, labels = toss.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('The Toss Decisions By Teams In The World Cup:')
plt.show()
```

### So in 17 matches, the teams decided to bat first, and in 13 matches, the teams chose to field first.

## The Top Scorers In The T20 World cup 2022:

```python
plt.bar(data["top scorer"], data["highest score"],width = 0.5, color ='maroon')
plt.title("The Top Scorers In The T20 World cup 2022")
plt.xlabel("Top Scorer")
plt.ylabel("highest Score")
plt.xlim(0,25)
plt.ylim(0,250)
plt.legend()
plt.grid()
plt.show()
```
